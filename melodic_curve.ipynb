```python
from music21 import note, interval, stream, converter

def iterate_score(score):
    # Dictionary to store the results
    score_data = {
        'title': score.metadata.title if score.metadata else "Untitled",
        'parts': {}
    }

    #print(score_data['title'])  # Print the title
    
    for part in score.parts:
        # Filter out Chord objects and get only Note objects
        notes = [n for n in part.recurse().notes if isinstance(n, note.Note)]
        
        part_data = {
            'highest_note': None,
            'lowest_note': None,
            'first_note_to_highest_interval': None,
            'last_note_to_highest_interval': None,
            'first_note_to_lowest_interval': None,
            'last_note_to_lowest_interval': None
        }

        if notes:  # Ensure there are notes to process
            highest_note = max(notes, key=lambda n: n.pitch.midi)
            lowest_note = min(notes, key=lambda n: n.pitch.midi)

            # Get the first and last notes
            first_note = notes[0]
            last_note = notes[-1]

            # Calculate the interval between the first/last note and the highest/lowest note
            first_note_to_highest_interval = interval.Interval(noteStart=first_note, noteEnd=highest_note)
            last_note_to_highest_interval = interval.Interval(noteStart=last_note, noteEnd=highest_note)
            first_note_to_lowest_interval = interval.Interval(noteStart=first_note, noteEnd=lowest_note)
            last_note_to_lowest_interval = interval.Interval(noteStart=last_note, noteEnd=lowest_note)

            # Store data in dictionary
            part_data['highest_note'] = highest_note.nameWithOctave
            part_data['lowest_note'] = lowest_note.nameWithOctave
            part_data['first_note_to_highest_interval'] = first_note_to_highest_interval.name
            part_data['last_note_to_highest_interval'] = last_note_to_highest_interval.name
            part_data['first_note_to_lowest_interval'] = first_note_to_lowest_interval.name
            part_data['last_note_to_lowest_interval'] = last_note_to_lowest_interval.name

            # Print the interval names
#             print(part.id)
#             print(f"First note interval from the highest note: {first_note_to_highest_interval.name}")
#             print(f"Last note interval from the highest note: {last_note_to_highest_interval.name}")
#             print(f"First note interval from the lowest note: {first_note_to_lowest_interval.name}")
#             print(f"Last note interval from the lowest note: {last_note_to_lowest_interval.name}")
        else:
            print(f"Part {part.id} contains no notes.")
        
        # Store part data using part ID as key
        score_data['parts'][part.id] = part_data

    return score_data

# Example usage
# Load a score using music21
score_example = converter.parse('inven01.krn')  # Adjust path as necessary
all_scores_data = [iterate_score(score) for score in scores ]
all_scores_data
```

```python
import matplotlib.pyplot as plt
from collections import defaultdict

# Prepare data for histograms
categories = [
    'first_note_to_highest_interval',
    'last_note_to_highest_interval',
    'first_note_to_lowest_interval',
    'last_note_to_lowest_interval'
]

# Initialize dictionary to store occurrences
category_occurrences = {part: {category: defaultdict(int) for category in categories} for part in ['spine_1', 'spine_0']}

# Sample data population
for entry in all_scores_data:
    parts = entry['parts']
    for part_name in ['spine_1', 'spine_0']:
        part_data = parts.get(part_name, {})
        for category in categories:
            interval = part_data.get(category, None)
            if interval:
                category_occurrences[part_name][category][interval] += 1

# Define the plotting function
def plot_histogram(data, title, ax):
    # Sorting the intervals by their numeric component
    keys = sorted(data.keys(), key=lambda x: (int(''.join(filter(str.isdigit, x))), x))
    values = [data[k] for k in keys]
    ax.bar(keys, values, color='skyblue')
    ax.set_title(title)
    ax.set_ylabel('Frequency')
    ax.set_xlabel('Interval')
    plt.setp(ax.get_xticklabels(), rotation=45, horizontalalignment='right')

# Setup for subplots
fig, axs = plt.subplots(nrows=8, figsize=(10, 20))  # 8 histograms

# Plotting all histograms
axes_idx = 0
for part_name in ['spine_1', 'spine_0']:
    for category in categories:
        plot_histogram(category_occurrences[part_name][category], f'{category} in {part_name}', axs[axes_idx])
        axes_idx += 1

# Tight layout to prevent overlap
plt.tight_layout()
plt.show()
```
